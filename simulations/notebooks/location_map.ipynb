```python
import os
import pyslim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
from tqdm.auto import tqdm
from sklearn.neighbors import KernelDensity
```

functions

```python
def get_shared_times(ts):
    'get shared evolutionary times between all samples in treesequence using general_stat'

    k = ts.num_samples
    W = np.identity(k) # each node, i in [0,1,...,k] given row vector of weights with 1 in column i and 0's elsewhere 
    def f(x): return (x.reshape(-1,1) * x).flatten() # matrix with 1's where branch above a node with value x contributes to shared time between samples 
    
    return ts.general_stat(
               W, f, k**2, mode='branch', windows='trees', polarised=True, strict=False
           ).reshape(ts.num_trees, k, k)
```

```python
def mle_dispersal(locations, shared_time):
    'maximum likelihood dispersal rate'

    not_missing = np.argwhere(np.isfinite(locations).any(axis=1)).flatten()
    if len(not_missing) < len(locations):
        print('removing', len(locations) - len(not_missing), 'samples with missing locations from dispersal estimate')
        locations = locations[not_missing]
        shared_time = shared_time[not_missing][:,not_missing]

    n = len(locations) #number of samples

    Tmat = np.identity(n) - [[1/n for _ in range(n)] for _ in range(n)]; Tmat = Tmat[:-1]; #matrix for mean centering and dropping one sample
    locs = np.matmul(Tmat, locations) #mean center locations and drop last sample (bc lost a degree of freedom when taking mean); see Lee & Coop 2017 Genetics eq A.16; note numpy broadcasts across the columns in locs
    stime = np.matmul(Tmat, np.matmul(shared_time, np.transpose(Tmat))) #mean center the shared time matrix

    Tinv = np.linalg.pinv(np.array(stime)) #inverse of shared time matrix

    SigmaHat = np.matmul(np.matmul(np.transpose(locs), Tinv), locs) / (n-1) #mle dispersal rate 
    
    return SigmaHat
```

```python
def get_mle_ancestral_locations(loci, focal_nodes, times, shared_times, locations, SIGMAs):
    'get the mle location, and covariance around it, of focal_nodes ancestor time generations ago'
    'shared_time is matrix of shared times between samples'
    'locations is matrix of locations of samples'
    'SIGMA is estimate of dispersal rate (covariance matrix)'

    mles = np.empty((len(loci),len(focal_nodes),len(times),2))
    covs = np.empty((len(loci),len(focal_nodes),len(times),2,2))
    
    for i,locus in enumerate(loci):
        shared_time = shared_times[locus]
    
        for j,focal_node in enumerate(focal_nodes):
            
            for t,time in enumerate(times):
    
                # shared times with ancestor
                tmrca = np.max(shared_time[focal_node]) #time to mrca
                tAA = tmrca - time #time to mrca from ancestor (ie shared time with itself)
                Atimes = [] #empty vector for shared times with other nodes
                for st in shared_time[focal_node]: #for each time the focal node shares with all other nodes
                    Atimes.append(np.min([tAA, st])) #the shared time with the ancestor of the focal node is the smaller of tAA and the time shared with the focal node
                Atimes = np.array(Atimes)

                # combine ancestor times with shared times among samples
                n,m = shared_time.shape #number of samples
                stime = np.zeros((n+1,m+1))
                stime[0,0] = tAA #first diagonal element
                stime[0,1:] = Atimes #remainder of first row
                stime[1:,0] = Atimes #remainder of first column
                stime[1:,1:] = shared_time #remainder of matrix

                # retrieve submatrices from combined times
                Sigma11 = stime[:1,:1]
                Sigma12 = stime[:1,1:]
                Sigma21 = stime[1:,:1]
                Sigma22 = stime[1:,1:]

                # mean centering
                Tmat = np.identity(n) - [[1/n for _ in range(n)] for _ in range(n)]; Tmat = Tmat[:-1] #mean centering matrix

                Sigma22c = np.matmul(Tmat, np.matmul(Sigma22, np.transpose(Tmat))) #mean centered covariance matrix of samples
                if len(Sigma22c) == 0:
                    Sigma22c = np.zeros((1,1)) #in case there is only one sample, which is removed by mean centering, just set shared time with self to be zero

                locationsc = np.matmul(Tmat, locations) #mean centered locations
                if len(locationsc) == 0:
                    locationsc = np.zeros((1,m)) #again, if we dropped the only sample just set to zero

                x = np.array([[1] + [-1/n] * n]) #vector to mean center ancestors shared time with itself
                Sigma11c = np.matmul(np.matmul(x, stime) , x.transpose()) # mean center ancestors shared time with itself

                Sigma21c = np.matmul(Tmat, Sigma21) - np.matmul(np.matmul(Tmat, Sigma22), np.ones(n).reshape(-1,1)/n) #mean center ancestors shared time with samples
                if len(Sigma21c) == 0:
                    Sigma21c = np.zeros((1,1)) #in case there is only one sample, which is removed by mean centering, just set shared time with ancestor to be zero

                # MLE ancestor location and covariance around it
                Sigma22c_pinv = np.linalg.pinv(Sigma22c)
                mles[i,j,t] = np.mean(locations, axis=0) + np.matmul( np.matmul(Sigma21c.transpose(), Sigma22c_pinv), locationsc)
                covs[i,j,t] = (Sigma11c - np.matmul(np.matmul(Sigma21c.transpose(), Sigma22c_pinv), Sigma21c)) * SIGMA

    return mles, covs
```

```python
def get_true_ancestral_locations(treeseq, focal_nodes, loci, times):
    'get ancestor locations for samples at times and loci given a treesequence with ancient individuals with location data'

    ancient_nodes = []
    # get all ancient nodes
    for time in times:

        # get all ancient nodes at given time
        ancient_nodes_time = []
        for ind in treeseq.individuals_alive_at(time): # for all individuals alive at this previous time

            # get both genomes
            for genome in treeseq.individual(ind).nodes:
                ancient_nodes_time.append(genome)

        # add list of ancient nodes at time t to larger list
        ancient_nodes.append(ancient_nodes_time)

    # loop over loci/trees
    ancestor_locs = []
    progress_bar = tqdm(total=len(loci)*len(focal_nodes)*len(times))
    for locus,tree in enumerate(treeseq.trees()):

        if locus > max(loci):
            break #stop if can
        elif locus in loci:

            # determine ancestor locations
            ancestor_locs_locus = []
            for ancient_nodes_time in ancient_nodes: #for each time

                ancestor_locs_locus_time = []
                for focal_node in focal_nodes: #for each focal node

                    for ancient_node in ancient_nodes_time: #for each ancient node

                        if tree.is_descendant(focal_node, ancient_node): #if ancient node is ancestor of focal node
                            ancestor_locs_locus_time_node = list(treeseq.individual_locations[treeseq.node(ancient_node).individual])[:2] #add location                            

                    progress_bar.update()

                    ancestor_locs_locus_time.append(ancestor_locs_locus_time_node)
                ancestor_locs_locus.append(ancestor_locs_locus_time)
            ancestor_locs.append(ancestor_locs_locus)

    progress_bar.close(); del progress_bar

    return np.array(ancestor_locs, dtype=object).swapaxes(1,2)
```

```python
def _plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma error ellipse based on the specified covariance
    matrix (`cov`). Additional keyword arguments are passed on to the 
    ellipse patch artist.

    Parameters
    ----------
        cov : The 2x2 covariance matrix to base the ellipse on
        pos : The location of the center of the ellipse. Expects a 2-element
            sequence of [x0, y0].
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the 
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    def _eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = _eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = patches.Ellipse(xy=pos, width=width, height=height, angle=theta, **kwargs)

    ax.add_artist(ellip)
    return ellip
```

```python
def ancient_plot(locations, focal_nodes, times, mles, covs, truths, title=None, fname=None, plot_confidence_intervals=True, plot_mles=False, plot_truths=True, mle_color=None, mle_alpha=0.1, mle_marker='^', truth_color=None, truth_alpha=0.1, truth_marker='x', plot_truth_focal_connector=True):

    # color map to plot extant nodes with color indicating location
    xmin = min(locations[:,0]) #minimum x location
    xmax = max(locations[:,0]) #maximum x location
    ymin = min(locations[:,1]) #minimum y location
    ymax = max(locations[:,1]) #maximum y location
    cmap = lambda x,y : ((x-xmin)/(xmax-xmin), 0, (y-ymin)/(ymax-ymin)) #scale red and blue values to be between 0 and 1 based on x and y location, respectively (and set gree to 0)
    colors = [cmap(loc[0],loc[1]) for loc in locations] #determine color of each chromosome
    bigcmap = lambda x,y : (255*(x-xmin)/(xmax-xmin), 0, 255*(y-ymin)/(ymax-ymin)) #scale bw 0 and 255 for alternative scale (svg tree plot)
    bigcolors  = [bigcmap(loc[0],loc[1]) for loc in locations]
    coldict = dict(zip([i for i in range(len(locations))], ["#%02x%02x%02x" % tuple(np.rint(i).astype('int')) for i in bigcolors])) #colors as dictionary, for tree plot

    # set up figure
    fig = plt.figure()
    ax = fig.add_subplot(111, aspect='equal')

    # plot locations of extant nodes
    ax.scatter(locations[:,0], locations[:,1], alpha=0.5, color=colors, zorder=0)

    for i,locus in enumerate(mles):
        for j,node in enumerate(locus):
            for t,time in enumerate(node):

                focal_node = focal_nodes[j] #get position of focal_node in locations array

                # plot confidence intervals of predicted ancestor location    
                if plot_confidence_intervals:
                    mean = list(mles[i][j][t])
                    cov = covs[i][j][t]
                    _plot_cov_ellipse(cov, mean, nstd=1.96, fill=False, color=colors[focal_node])
                
                # plot mle ancestor locations
                if plot_mles:
                    mean = mles[i][j][t]
                    if mle_color is None:
                        color = colors[focal_node]
                    else:
                        color = mle_color
                    ax.scatter(mean[0], mean[1], color=color, alpha=mle_alpha, marker=mle_marker, zorder=1)

                # plot true ancestor locations
                if plot_truths:
                    truth = truths[i][j][t]
                    if truth_color is None:
                        color = colors[focal_node]
                    else:
                        color = truth_color
                    ax.scatter(truth[0], truth[1], color = color, alpha=truth_alpha, marker=truth_marker, zorder=2)
                if plot_truth_focal_connector:
                    ax.plot([truth[0],locations[focal_node,0]], [truth[1],locations[focal_node,1]], color = colors[focal_node]) #line connect truth to focal node

    # pretty-up the axes
    ax.set_xlabel('x')
    ax.set_ylabel('y')
#     labels = [item.get_text() for item in ax.get_xticklabels()]
#     empty_string_labels = ['']*len(labels)
#     ax.set_xticklabels(empty_string_labels)
#     labels = [item.get_text() for item in ax.get_yticklabels()]
#     empty_string_labels = ['']*len(labels)
#     ax.set_yticklabels(empty_string_labels)

    if title != None:
        plt.title(title)

    if fname != None:
        plt.tight_layout()
        plt.savefig(fname)

    plt.show()
```

```python
def kde2D(x, y, mles, ancestors=None, bandwidth=1.0, xbins=100j, ybins=100j, nlevels=100, cmap=plt.cm.Blues, figsize=(5,5), title_text=None, fname = None, **kwargs): 
    """Build 2D kernel density estimate (KDE)."""

    # create grid of sample locations (default: 100x100)
    xx, yy = np.mgrid[x.min():x.max():xbins, 
                      y.min():y.max():ybins]
    xy_sample = np.vstack([yy.ravel(), xx.ravel()]).T

    # fit density kernel
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    xy_train  = np.vstack([mles[:,1], mles[:,0]]).T
    kde_skl.fit(xy_train)

    # score_samples() returns the log-likelihood of the samples
    z = np.exp(kde_skl.score_samples(xy_sample))
    z = np.reshape(z, xx.shape)
    
    fig = plt.figure(figsize=figsize)
    plt.subplot(1, 1, 1)
    
    # plot contours of the density
    levels = np.linspace(0, z.max(), nlevels)
    plt.contourf(xx, yy, z, levels=levels, cmap=cmap)
    
    #plot true locations of ancestors
    if ancestors is not None:
        plt.scatter(ancestors[:,0], ancestors[:,1], facecolor='k', marker = 'x', alpha=0.1)
    
    # pretty up
    plt.xlim([x.min(),x.max()])
    plt.ylim([y.min(),y.max()])
    plt.tight_layout()
    if title_text is not None:
        plt.title(title_text)

    if fname is not None:
        plt.savefig(fname)

    plt.show()
```

```python
L = 1e8 #number of sites to simulate (length of chromosome)
RBP = 5e-9 #recombination rate per site
LAMBDA = 2 #number of offspring per pair per generation when no competition
K = 2 #carrying capacity per unit area
W = 50 #width and height of area simulated
SIGMAi = 0.5 #SD of competition and mating kernels
SIGMAd = 0.5 #SD of dispersal kernel
N = W**2 * K #equil popn size
MAXT = 4*N #number of generations to run the simulation for
anc_gens = range(1,1001,10) #gens to record ancestral individuals at
k = 50 #number of diploid samples
rep = 0
Ne = 5000
U = 1.25e-8
d = 100

datadir = '../data/'
filename = datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_true.trees' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d)
plotdir = '../plots/'
```

## load tree sequence and get locations

```python
ts = pyslim.load(filename)
ts
```

```python
locations = []
for i in ts.samples():
    ind = ts.node(i).individual #individual each sample belongs to
    loc = ts.individual_locations[ind] #location of each individual
    locations.append(loc[:2]) #x,y location of each sample
locations = np.array(locations)
```

## coalescing confidence intervals

get shared times at a locus

```python
nloci = 1 #number of loci to get shared times at
keep_loci = np.linspace(0, ts.num_trees-1, nloci).round().astype(int) #uniformly sample from tree sequence
breakpoints = np.array([i for i in ts.breakpoints()]) #breakpoints between trees
intervals = [[i,i+1] for i in breakpoints[keep_loci]] #the first and second site in each tree we want to keep
ts_keep = ts.keep_intervals(intervals) #keep just the intervals we want
shared_times = get_shared_times(ts_keep) #get the shared times
even_ix = [2*i for i in range(nloci)]
shared_times = shared_times[even_ix] #for some reason keep_intervals keeps the intermediate intervals too, so remove those by choosing just even trees
```

find some recent coalescences

```python
locus = 0 #choose an arbitrary locus
tmrca = np.max(shared_times[locus][0]) #tmrca
ctimes = tmrca - shared_times[locus] #coalescent times

# find nodes that coalesce recently
x,y = np.where((ctimes < 500) & (ctimes > 0))

# order by distance between nodes
d0 = []
for i,j in zip(x,y):
    d0.append(np.sum((locations[i] - locations[j])**2)**0.5) #distance between coalescing nodes
dix = np.argsort(d0) #order distances in ascending order
```

get mle and true locations for this example

```python
loci = [locus]
i = dix[-20] #choose interesting long distance recent dispersal
focal_nodes = [x[i],y[i]] #the two samples
times = range(100,600,100) #times to plot

# mle dispersal rates
SIGMAs=[]
for i,locus in enumerate(loci):
    SIGMA = mle_dispersal(locations=locations, shared_time=shared_times[i])
    SIGMAs.append(SIGMA)

# mle locations
mles, covs = get_mle_ancestral_locations(loci=loci, focal_nodes=focal_nodes, times=times, shared_times=shared_times, locations=locations, SIGMAs=SIGMAs)

# true locations
truths = get_true_ancestral_locations(treeseq=ts, focal_nodes=focal_nodes, loci=loci, times=times)
```
