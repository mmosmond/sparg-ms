```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def barplot(SIGMAds, mles_trues, mles_infs, mcle_trues, mcle_infs, fname=None):
    
    #params
    cmap = plt.get_cmap('tab10')
    labels = [r'$\sigma_x$',r'$\sigma_y$',r'$\rho$']
    x = np.array([100*(1-(1-2*RBP)**i)/2 for i in which_sites]) #convert sites to cM
    truths = np.array([[i,i,0] for i in SIGMAds])
    ymins = [0, 0, -1]
    ymaxs = [max(SIGMAds)+min(SIGMAds), max(SIGMAds)+min(SIGMAds), 1]
    dj = 0.1
    alpha = 0.75
    fontsize = 16
    legendfontsize = 14
    ticklabelfontsize = 12
    j = 0 #which SIGMAd to show genome wide
    rep = 0 #which rep to show genome wide

    #set up
    fig = plt.figure(figsize=(10, 10))
    ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
    ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
    ax3 = plt.subplot2grid((3, 3), (2, 0), colspan=2)
    ax4 = plt.subplot2grid((3, 3), (0, 2))
    ax5 = plt.subplot2grid((3, 3), (1, 2))
    ax6 = plt.subplot2grid((3, 3), (2, 2))

    #mles across genome
    for i,ax in enumerate([ax1,ax2,ax3]):
        ax.plot(x, [truths[j,i] for _ in x], color='k', linestyle='-', label='truth')
        ax.plot(x, mles_trues[j,rep,:,i], color = cmap(0), alpha = 1, label='true trees')
        ax.plot(x, [mcle_trues[j,rep,i] for _ in x], color = cmap(0), alpha = alpha)
        ax.plot(x, mles_infs[j,rep,:,i], color = cmap(1), alpha = 1, label='inferred trees')
        ax.plot(x, [mcle_infs[j,rep,i] for _ in x], color = cmap(1), alpha = alpha)
        ax.set_ylim(ymins[i], ymaxs[i])
        ax.set_ylabel(labels[i], fontsize=fontsize)
        ax.tick_params(axis='both',labelsize=ticklabelfontsize)

    # pretty-up
    ax1.legend(ncol=3, loc='upper center', fontsize=legendfontsize)
    ax.set_xlabel('genomic position (cM)', fontsize=fontsize)
    ax1.set_xticklabels([])
    ax2.set_xticklabels([])
    ax.set_yticks(np.linspace(ymins[i], ymaxs[i],5))

    #mcles
    for i,ax in enumerate([ax4,ax5,ax6]):
        for j,truth in enumerate(truths[:,i]):
            ax.plot([j-dj,j+dj], [truth,truth], color='k') #truth
            ax.scatter([j] * NREPS, mcle_trues[j,:,i], color = cmap(0), alpha = alpha) #mcle from true trees
            ax.scatter([j] * NREPS, mcle_infs[j,:,i], color = cmap(1), alpha = alpha) #mcle from inferred trees        
        #pretty-up    
        ax.set_ylim(ymins[i], ymaxs[i])
        ax.set_xticks(range(len(truths)))
        ax.set_xticklabels([])
        ax.set_yticklabels([])
#     ax.set_xticklabels(['A','B','C'], fontsize=ticklabelfontsize)
    ax.set_xlabel('parameter combo', fontsize=fontsize)
    ax.set_yticks(np.linspace(ymins[i], ymaxs[i],5))

    plt.tight_layout()
    if fname != None:
        plt.savefig(fname)
    plt.show()
```

```python
def corrplot(SIGMAds, mles_trues, mles_infs, mcle_trues, mcle_infs, fname=None):

    # #params
    cmap = plt.get_cmap('tab10')
    labels = [r'inferred trees $\sigma_x$',r'inferred trees $\sigma_y$',r'inferred trees $\rho$']
    truths = np.array([[i,i,0] for i in SIGMAds])
    j = 0 #which SIGMAd to show genome wide
    ds = 0.5
    # ymins = [SIGMAds[j]-ds, SIGMAds[j]-ds, -1]
    # ymaxs = [SIGMAds[j]+ds, SIGMAds[j]+ds, 1]
    ymins = [min(truths[:,i])-ds for i in range(3)]
    ymaxs = [max(truths[:,i])+ds for i in range(3)]
    # dj = 0.1
    alpha = 0.75
    fontsize = 16
    legendfontsize = 12
    ticklabelfontsize = 12

    rep = 0 #which rep to show genome wide

    # #set up
    fig = plt.figure(figsize=(10, 10))
    ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=1)
    ax2 = plt.subplot2grid((3, 2), (1, 0), colspan=1)
    ax3 = plt.subplot2grid((3, 2), (2, 0), colspan=1)
    ax4 = plt.subplot2grid((3, 2), (0, 1))
    ax5 = plt.subplot2grid((3, 2), (1, 1))
    ax6 = plt.subplot2grid((3, 2), (2, 1))

    # #mles across genome
    for i,ax in enumerate([ax1,ax2,ax3]):
        ax.plot([ymins[i],ymaxs[i]],[ymins[i],ymaxs[i]], color='k', zorder=0)
        ax.scatter(mles_trues[j,rep,:,i], mles_infs[j,rep,:,i], color = 'gray', alpha = alpha, zorder=1, label='loci mles')
        ax.scatter(mcle_trues[j,rep,i], mcle_infs[j,rep,i], color = cmap(j), alpha = 1, zorder=2, label='mcle')
        ax.scatter(truths[j,i], truths[j,i], color=cmap(j), zorder=3, label='truth', marker='x')
        ax.set_xlim(ymins[i], ymaxs[i])
        ax.set_ylim(ymins[i], ymaxs[i])
        ax.set_ylabel(labels[i], fontsize=fontsize)
        ax.tick_params(axis='both',labelsize=ticklabelfontsize)

    # # pretty-up
    ax1.legend(ncol=3, loc='upper center', fontsize=legendfontsize)
    ax.set_xlabel('true trees', fontsize=fontsize)
    # ax1.set_xticklabels([])
    # ax2.set_xticklabels([])
    # ax.set_yticks(np.linspace(ymins[i], ymaxs[i], 5))

    # #mcles
    mins = [min(truths[:,i])-ds for i in range(3)]
    maxs = [max(truths[:,i])+ds for i in range(3)]
    for i,ax in enumerate([ax4,ax5,ax6]):
        for j,truth in enumerate(truths):
            for n in range(NREPS):
                ax.scatter(mcle_trues[j,n,i], mcle_infs[j,n,i], color = cmap(j), alpha = alpha, zorder=2)
            ax.scatter(truth[i], truth[i], color=cmap(j), zorder=3, marker='x')
        ax.set_xlim(mins[i], maxs[i])
        ax.set_ylim(mins[i], maxs[i])
        ax.plot([mins[i],maxs[i]],[mins[i],maxs[i]], color='k')
    #     ax.set_ylabel(labels[i], fontsize=fontsize)
        ax.tick_params(axis='both',labelsize=ticklabelfontsize)

    # # pretty-up
    # ax4.legend(ncol=3, loc='upper center', fontsize=legendfontsize)
    ax.set_xlabel('true trees', fontsize=fontsize)
    # ax1.set_xticklabels([])
    # ax2.set_xticklabels([])
    # ax.set_yticks(np.linspace(ymins[i], ymaxs[i], 5))

    plt.tight_layout()
    if fname != None:
        plt.savefig(fname)
    plt.show()
```

```python
def corrxplot(truths, mcle_trues, mcle_infs, i=0, label='$\sigma_x$', ymin=None, ymax=None, ticks=None, 
              fname=None):

    # params
    cmap = plt.get_cmap('viridis')
    labels = [r'inferred trees %s' %label]
    ds = 0.1 #padding
    if ymin is None:
        ymin = min(truths) - ds
    if ymax is None:
        ymax = max(truths) + 4*ds
    if ticks is None:
        ticks = truths
    alpha = 0.75
    fontsize = 16
    legendfontsize = 12
    ticklabelfontsize = 12

    # set up
    fig = plt.figure(figsize=(5, 5))
    ax1 = plt.subplot(1,1,1)

    # plot
    for j,truth in enumerate(truths):
        for n in range(NREPS):
            ax1.scatter(mcle_trues[j,n,i], mcle_infs[j,n,i], color = cmap(j/len(truths)), alpha = alpha, zorder=2) #mcle of inf vs true tree (sigmax only)
        ax1.scatter(truth, truth, color=cmap(j/len(truths)), zorder=3, marker='X') #simulated parameter value
    ax1.plot([ymin,ymax],[ymin,ymax], color='gray', zorder=1) #1:1 line

    # pretty-up
    ax1.set_xlim(ymin, ymax)
    ax1.set_ylim(ymin, ymax)
    ax1.set_ylabel(labels[0], fontsize=fontsize)
    ax1.tick_params(axis='both',labelsize=ticklabelfontsize)
    ax1.set_xlabel(r'true trees %s' %label, fontsize=fontsize)
    ax1.xaxis.set_major_locator(ticker.FixedLocator(ticks))
    ax1.yaxis.set_major_locator(ticker.FixedLocator(ticks))
    
    plt.tight_layout()
    if fname != None:
        plt.savefig(fname)
    plt.show()
```

```python
def barxplot(SIGMAd, mles_trues, mles_infs, mcle_trues, mcle_infs, fname=None):
    
    #params
    cmap = plt.get_cmap('tab10')
    label = r'$\sigma_x$'
#     x = np.array([100*(1-(1-2*RBP)**i)/2 for i in which_sites]) #convert sites to cM
    x = which_sites
    truth = SIGMAd
    ymin = 0.25
    ymax = 1
    dj = 0.1
    alpha = 0.75
    fontsize = 16
    legendfontsize = 14
    ticklabelfontsize = 12
    reps = range(10) #which rep to show genome wide

    #set up
    fig = plt.figure(figsize=(10, 5))
    ax1 = plt.subplot(1,1,1)

    #mles across genome
    i = 0 #sigmax only
    labeled = False
    for rep in reps:
        if not labeled:
            labels = ['truth', 'true trees', 'inferred trees']
            labeled = True
        else:
            labels = [None] * 3
        ax1.plot(x, [truth for _ in x], color='k', linestyle='-', label=labels[0])
        ax1.plot(x, mles_trues[rep,:,i], color = cmap(0), alpha = 0.25)
        ax1.plot(x, [mcle_trues[rep,i] for _ in x], color = cmap(0), alpha = 0.5, label=labels[1])
        ax1.plot(x, mles_infs[rep,:,i], color = cmap(1), alpha = 0.25) #, marker='^')
        ax1.plot(x, [mcle_infs[rep,i] for _ in x], color = cmap(1), alpha = 0.5, label=labels[2]) #, linestyle='--')

    # pretty-up
    ax1.set_xlim(0, L)
    ax1.set_ylim(ymin, ymax)
    ax1.set_ylabel(label, fontsize=fontsize)
    ax1.tick_params(axis='both', labelsize=ticklabelfontsize)
    ax1.legend(ncol=3, loc='upper center', fontsize=legendfontsize)
    ax1.set_xlabel('genomic position (site)', fontsize=fontsize)
#     ax1.set_xticklabels([])
#     ax1.set_yticks(np.linspace(ymin, ymax, 4))
    ax1.yaxis.set_major_locator(ticker.FixedLocator([0.25,0.5,0.75]))

    plt.tight_layout()
    if fname != None:
        plt.savefig(fname)
    plt.show()
```

parameters

```python
datadir = '../data/'
plotdir = '../plots/'

#params to get files
L = 100000000 #sequence length
RBP = 5e-9 #recombination rate
LAMBDA = 2.0 #birth rate
K = 2.0 #carrying capacity
W = 50 #width of habitat
SIGMAi = 0.5 #interaction kernel SD
MAXT = 20000 #num gens of spatial sim
NREPS = 10 #number of reps
Ne = 5000 #effective pop size estimate
U = 1.25e-8 #mutation rate
k = 50 #number of sampled individuals
d = 100 #max distance from center to sample individuals
nloci = 100 # the number of loci to estimate sigma at
M = 10 #number of samples of tree
numiter = 5 #number of relate iterations

#params for plotting
nsites = L-1e3 # total number of sites in chromosome, note we subtract some off because the inferred chromosomes are sometimes shorter
which_sites = np.linspace(0, nsites, nloci, dtype=int) # sites to get trees at

SIGMAds = [0.5,0.75,1,1.25,1.5,1.75,2]
```

```python
tCutoff = 10000 #gens to go back to
```

```python
def true_mle_pattern(datadir, SIGMAd, rep, locus):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dlocus_%dtCutoff_mle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,locus,tCutoff)

def inf_mle_pattern(datadir, SIGMAd, rep, locus):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dlocus_%dtCutoff_%dM_mle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M)

def true_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff)

def inf_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M)
```

```python
def get_mles(SIGMAds):
    
    mles_trues = np.empty((len(SIGMAds),NREPS,nloci,4))
    mles_infs = np.empty((len(SIGMAds),NREPS,nloci,4))
    
    for i,SIGMAd in enumerate(SIGMAds):
        
        for rep in range(NREPS):

            for locus in range(nloci):
                
                filename = true_mle_pattern(datadir, SIGMAd, rep, locus)
                try:
                    file = np.load(filename, allow_pickle=True).item()
                    if file.success == True:
                        mles_trues[i,rep,locus] = file.x
                    else:
                        print('SIGMAd %f rep %d locus %d with true trees failed!' %(SIGMAd,rep,locus))
                        mles_trues[i,rep,locus] = [np.nan] * 4
                except:
                    print('SIGMAd %f rep %d locus %d with true trees missing!' %(SIGMAd,rep,locus))
                    mles_trues[i,rep,locus] = [np.nan] * 4

                filename = inf_mle_pattern(datadir, SIGMAd, rep, locus)
                try:
                    file = np.load(filename, allow_pickle=True).item()
                    if file.success == True:
                        mles_infs[i,rep,locus] = file.x
                    else:
                        print('SIGMAd %f rep %d locus %d with inferred trees failed!' %(SIGMAd,rep,locus))
                        mles_infs[i,rep,locus] = [np.nan] * 4
                except:
                    print('SIGMAd %f rep %d locus %d with inferred trees missing!' %(SIGMAd,rep,locus))
                    mles_infs[i,rep,locus] = [np.nan] * 4
                    
    return mles_trues, mles_infs
```

```python
def get_mcles(SIGMAds):
    
    mcles_trues = np.empty((len(SIGMAds),NREPS,4))
    mcles_infs = np.empty((len(SIGMAds),NREPS,4))
    
    for i,SIGMAd in enumerate(SIGMAds):
        
        for rep in range(NREPS):
            
            filename = true_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_trues[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with true trees failed!' %(SIGMAd,rep))
                    mcles_trues[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with true trees missing!' %(SIGMAd,rep))
                mcles_trues[i,rep] = [np.nan] * 4                
                    
            filename = inf_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_infs[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with inferred trees failed!' %(SIGMAd,rep))
                    mcles_infs[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with inferred trees missing!' %(SIGMAd,rep))
                mcles_infs[i,rep] = [np.nan] * 4     
                
    return mcles_trues, mcles_infs
```

```python
def true_mle_pattern(datadir, SIGMAd, rep, locus):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dlocus_%dtCutoff_mle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,locus,tCutoff)

def inf_mle_pattern(datadir, SIGMAd, rep, locus):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dlocus_%dtCutoff_%dM_mle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M)

def true_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff)

def inf_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M)
```

```python
def get_mcles(SIGMAds):
    
    mcles_trues = np.empty((len(SIGMAds),NREPS,4))
    mcles_infs = np.empty((len(SIGMAds),NREPS,4))
    
    for i,SIGMAd in enumerate(SIGMAds):
        
        for rep in range(NREPS):
            
            filename = true_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_trues[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with true trees failed!' %(SIGMAd,rep))
                    mcles_trues[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with true trees missing!' %(SIGMAd,rep))
                mcles_trues[i,rep] = [np.nan] * 4                
                    
            filename = inf_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_infs[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with inferred trees failed!' %(SIGMAd,rep))
                    mcles_infs[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with inferred trees missing!' %(SIGMAd,rep))
                mcles_infs[i,rep] = [np.nan] * 4     
                
    return mcles_trues, mcles_infs
```

## compare with and without importance sampling

```python
tCutoff = 100
```

```python
def true_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff)

def inf_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M)

def unimp_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mcle_dispersal_unimp.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff)

def unimp_mcle_pattern(datadir, SIGMAd, rep):
    return datadir + 'slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%d_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_dispersal_unimp.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M)
```

```python
def get_mcles(SIGMAds):
    
    mcles_infs = np.empty((len(SIGMAds),NREPS,4))
    mcles_unimps = np.empty((len(SIGMAds),NREPS,4))
    
    for i,SIGMAd in enumerate(SIGMAds):
        
        for rep in range(NREPS):
            
            filename = inf_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_infs[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with IS failed!' %(SIGMAd,rep))
                    mcles_infs[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with IS missing!' %(SIGMAd,rep))
                mcles_infs[i,rep] = [np.nan] * 4                
                    
            filename = unimp_mcle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mcles_unimps[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d without IS failed!' %(SIGMAd,rep))
                    mcles_unimps[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d without IS missing!' %(SIGMAd,rep))
                mcles_unimps[i,rep] = [np.nan] * 4     
                
    return mcles_infs, mcles_unimps
```

load data

```python
mcle_infs, mcle_unimps = get_mcles(SIGMAds)
```

plot

```python
def corrxplot(truths, mcle_trues, mcle_infs, i=0, xlabel = r'true trees $\sigma_x$', ylabel=r'inferred trees $\sigma_x$', ymin=None, ymax=None, ticks=None, 
              fname=None):

    # params
    cmap = plt.get_cmap('viridis')
    ds = 0.1 #padding
    if ymin is None:
        ymin = min(truths) - ds
    if ymax is None:
        ymax = max(truths) + 4*ds
    if ticks is None:
        ticks = truths
    alpha = 0.75
    fontsize = 16
    legendfontsize = 12
    ticklabelfontsize = 12

    # set up
    fig = plt.figure(figsize=(5, 5))
    ax1 = plt.subplot(1,1,1)

    # plot
    for j,truth in enumerate(truths):
        for n in range(NREPS):
            ax1.scatter(mcle_trues[j,n,i], mcle_infs[j,n,i], color = cmap(j/len(truths)), alpha = alpha, zorder=2) #mcle of inf vs true tree (sigmax only)
        ax1.scatter(truth, truth, color=cmap(j/len(truths)), zorder=3, marker='X') #simulated parameter value
    ax1.plot([ymin,ymax],[ymin,ymax], color='gray', zorder=1) #1:1 line

    # pretty-up
    ax1.set_xlim(ymin, ymax)
    ax1.set_ylim(ymin, ymax)
    ax1.set_ylabel(ylabel, fontsize=fontsize)
    ax1.tick_params(axis='both',labelsize=ticklabelfontsize)
    ax1.set_xlabel(xlabel, fontsize=fontsize)
    ax1.xaxis.set_major_locator(ticker.FixedLocator(ticks))
    ax1.yaxis.set_major_locator(ticker.FixedLocator(ticks))
    
    plt.tight_layout()
    if fname != None:
        plt.savefig(fname)
    plt.show()
```

```python
fname = plotdir + 'sim_sigmax_5numiter_%dtCutoff_unimp.pdf' %tCutoff
corrxplot(SIGMAds, mcle_infs, mcle_unimps, xlabel = r'with importance sampling $\sigma_x$', ylabel = r'without importance sampling $\sigma_x$', ymax=3.5, ticks = [0.5, 1, 1.5, 2, 2.5, 3, 3.5], fname=fname)
```
