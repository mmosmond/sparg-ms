# Plotting root mean squared error of ancestor locations

modules

```python
import pyslim
import matplotlib.pyplot as plt
from importlib import reload  
import sparg
sparg = reload(sparg)
import numpy as np
```

parameters

```python
datadir = '../data/'
plotdir = '../plots/'

L = 1e8 #number of sites to simulate (length of chromosome)
RBP = 5e-9 #recombination rate per site
LAMBDA = 2 #number of offspring per pair per generation when no competition
K = 2 #carrying capacity per unit area
W = 50 #width and height of area simulated
SIGMAi = 0.5 #SD of competition and mating kernels
SIGMAd = 0.5 #SD of dispersal kernel
N = W**2 * K #equil popn size
MAXT = 4*N #number of generations to run the simulation for
k = 50 #number of diploid samples
rep = 0
Ne = 5000
U = 1.25e-8
d = 100
nloci = 100
times = range(10,1001,10)
tCutoff=1e4
```

```python
numiter = 5
M = 10
tCutoff = 10000
NREPS = 1
loc_tCutoff = 10000
```

functions

```python
def true_mle_pattern(datadir, SIGMAd, rep):
    return datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff)

def inf_mle_pattern(datadir, SIGMAd, rep):
    return datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_dispersal.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M)
```

```python
def get_mles(SIGMAds,NREPS):
    
    mles_trues = np.empty((len(SIGMAds),NREPS,4))
    mles_infs = np.empty((len(SIGMAds),NREPS,4))
    
    for i,SIGMAd in enumerate(SIGMAds):
        
        for rep in range(NREPS):
                
            filename = true_mle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mles_trues[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with true trees failed!' %(SIGMAd,rep))
                    mles_trues[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with true trees missing!' %(SIGMAd,rep))
                mles_trues[i,rep] = [np.nan] * 4

            filename = inf_mle_pattern(datadir, SIGMAd, rep)
            try:
                file = np.load(filename, allow_pickle=True).item()
                if file.success == True:
                    mles_infs[i,rep] = file.x
                else:
                    print('SIGMAd %f rep %d with inferred trees failed!' %(SIGMAd,rep))
                    mles_infs[i,rep] = [np.nan] * 4
            except:
                print('SIGMAd %f rep %d with inferred trees missing!' %(SIGMAd,rep))
                mles_infs[i,rep] = [np.nan] * 4
                    
    return mles_trues, mles_infs
```

```python
def get_true_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,locus,tCutoff,loc_tCutoff):
    return datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_true_%dnloci_processed_%dtCutoff_mle-locs_%dlocus_%dloc-tCutoff.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,tCutoff,locus,loc_tCutoff)

def get_inf_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M,loc_tCutoff):
    return datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_MLEstyle-locs_%dlocus_%dloc-tCutoff.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M,locus,loc_tCutoff)

def get_inf_blup_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M,loc_tCutoff):
    return datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_%dnumiter_inferred_%dnloci_processed_%dtCutoff_%dM_mcle_BLUPstyle-locs_%dlocus_%dloc-tCutoff.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,tCutoff,M,locus,loc_tCutoff)
```

first check all mle dispersal rates found OK

```python
mles_trues, mles_infs = get_mles([SIGMAd],1)
```

```python
print(mles_trues, mles_infs)
```

now move on to locations...

get node order to align data

```python
treename = datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_true_simple.trees' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d)
ts = pyslim.load(treename)

node_list = []
for ind in ts.individuals():
    if ind.id in range(int(ts.num_samples/2)):
        node_list.append(ind.nodes)
node_list = np.array(node_list).flatten()
```

get mle locations from true trees

```python
mle_locs = []
for locus in range(nloci):
    file = get_true_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,nloci,locus,tCutoff,loc_tCutoff)
    try:
        mle = np.load(file)[0]
        mle_locs.append(mle)
    except:
        print('error')
true_mle_locs = np.array(mle_locs)
true_mle_locs = true_mle_locs[:,:,node_list] #put in same order as inferred
true_mle_locs = true_mle_locs.swapaxes(1,2) #order: locus, node, time
```

get mle locations from inferred trees

```python
mle_locs = []
for locus in range(nloci):
    file = get_inf_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M,loc_tCutoff)
    try:
        mle = np.load(file)[0]
        mle_locs.append(mle)
    except:
        print('error at locus %d' %locus)
inf_mle_locs = np.array(mle_locs).swapaxes(1,2) #order: locus, node, time
```

get mle locations from inferred trees using BLUP method

```python
mle_locs = []
for locus in range(nloci):
    file = get_inf_blup_filename(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d,numiter,nloci,locus,tCutoff,M,loc_tCutoff)
    try:
        mle = np.load(file)[0]
        mle_locs.append(mle)
    except:
        print('error at locus %d' %locus)
inf_blup_locs = np.array(mle_locs).swapaxes(1,2) #order: locus, node, time
```

get true locations of ancestors

```python
filename = datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_true_anc_locs.npy' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d)
true_locs = np.load(filename)
true_locs = true_locs[:,node_list,:,:2] #put in same order as inferred and drop 3rd dimension
```

get current locations

```python
filename = datadir + 'retain_slim_%dL_%.0eRBP_%.1fLAMBDA_%.1fK_%dW_%sSIGMAi_%sSIGMAd_%dMAXT_%dnrep_%dNe_%.2eU_%dk_%dd_sample_reordered.locs' %(L,RBP,LAMBDA,K,W,SIGMAi,SIGMAd,MAXT,rep,Ne,U,k,d)
locations = np.loadtxt(filename)[:,:2] #locations of contemporary nodes in correct order
today_locs = np.array([[locations for _ in range(len(times))] for _ in range(nloci)]).swapaxes(1,2) #repeat to make same shape as true locs: loci, nodes, times
```

check things look right

```python
cmap = plt.cm.tab10
label = ['mle (true trees)', 'mle (inferred trees)', 'blup (inferred trees)', 'truth']
l = 49
for t in [0,9,99]: #time indices
    for n in range(0,10,2): #range(1,100,2):
        for i,data in enumerate([true_mle_locs, inf_mle_locs, inf_blup_locs, true_locs]):
            plt.scatter(today_locs[l,n,t,0], today_locs[l,n,t,1], color='k', label='sample')
            plt.scatter(data[l,n,t,0], data[l,n,t,1], color=cmap(i), label=label[i])
            plt.plot([data[l,n,t,0], today_locs[l,n,t,0]], [data[l,n,t,1], today_locs[l,n,t,1]], color=cmap(i))
    plt.legend(label)
    plt.title('generation %d' %times[t])
    plt.xlim([0,50])
    plt.ylim([0,50])
    plt.show()
```

root mean squared errors of mles

```python
rmse_true = (np.nanmean(np.sum((true_mle_locs - true_locs)**2, axis=-1), axis=(0,1)))**0.5 #root mean (over loci and nodes) squared euclidean distance (error) between mle and truth at each time
rmse_inf = (np.nanmean(np.sum((inf_mle_locs - true_locs)**2, axis=-1), axis=(0,1)))**0.5 #root mean (over loci and nodes) squared euclidean distance (error) between mle and truth at each time
rmse_blup = (np.nanmean(np.sum((inf_blup_locs - true_locs)**2, axis=-1), axis=(0,1)))**0.5 #root mean (over loci and nodes) squared euclidean distance (error) between mle and truth at each time
```

the errors of some simple guesses

```python
# under hypothesis the ancestors don't move
rmse_today = (np.mean(np.sum((today_locs - true_locs)**2, axis=-1), axis=(0,1)))**0.5 #root mean (over loci and nodes) squared euclidean distance (error) between mle and truth at each time

# under hypothesis the ancestors are all at the mean location always 
mean_location = np.mean(locations, axis=0)
mean_locs = np.array([[[mean_location for _ in range(len(times))] for _ in range(len(locations))] for _ in range(nloci)]) #repeat to make same shape as true locs
rmse_mean = (np.mean(np.sum((mean_locs - true_locs)**2, axis=-1), axis=(0,1)))**0.5 #root mean (over loci and nodes) squared euclidean distance (error) between mle and truth at each time
```

plot

```python
fname = plotdir + 'rmse_5numiter.pdf'
cmap = plt.get_cmap('tab10')

plt.plot(times, rmse_mean, label='mean location', color='k')
plt.plot(times, rmse_today, label='current locations', color='gray')
# plt.plot(times, rmse_blup, label='blup (inferred trees)', color=cmap(1), linestyle='--')
plt.plot(times, rmse_inf, label='mle (inferred trees)', color=cmap(1))
plt.plot(times, rmse_true, label='mle (true trees)', color=cmap(0))

plt.xlabel('generations ago', fontsize=16)
plt.ylabel('root mean squared error', fontsize=16)

plt.tick_params(axis='both',labelsize=12)

plt.legend(fontsize=12)
plt.savefig(fname)
plt.tight_layout()

plt.show()
```

average over loci

```python
mean_true_locs = np.mean(true_locs, axis=0)
mean_true_mle_locs = np.mean(true_mle_locs, axis=0)
mean_inf_mle_locs = np.mean(inf_mle_locs, axis=0)
mean_inf_blup_locs = np.mean(inf_blup_locs, axis=0)
mean_today_locs = np.mean(today_locs, axis=0)
mean_mean_locs = np.mean(mean_locs, axis=0)
```

```python
rmse_mean_true = (np.nanmean(np.sum((mean_true_mle_locs - mean_true_locs)**2, axis=-1), axis=0))**0.5 #root mean (over nodes) squared euclidean distance (error) between mean mle and mean truth at each time
rmse_mean_inf = (np.nanmean(np.sum((mean_inf_mle_locs - mean_true_locs)**2, axis=-1), axis=0))**0.5 #root mean (over nodes) squared euclidean distance (error) between mean mle and mean truth at each time
rmse_mean_inf_blup = (np.nanmean(np.sum((mean_inf_blup_locs - mean_true_locs)**2, axis=-1), axis=0))**0.5 #root mean (over nodes) squared euclidean distance (error) between mean mle and mean truth at each time
rmse_mean_today = (np.mean(np.sum((mean_today_locs - mean_true_locs)**2, axis=-1), axis=0))**0.5 #root mean (over nodes) squared euclidean distance (error) between mean mle and mean truth at each time
rmse_mean_mean = (np.mean(np.sum((mean_mean_locs - mean_true_locs)**2, axis=-1), axis=0))**0.5 #root mean (over nodes) squared euclidean distance (error) between mean mle and mean truth at each time
```

plot

```python
fname = plotdir + 'rmse_mean_5numiter.pdf'

cmap = plt.get_cmap('tab10')

plt.plot(times, rmse_mean_mean, label='mean location', color='k')
plt.plot(times, rmse_mean_today, label='current locations', color='gray')
# plt.plot(times, rmse_mean_inf_blup, label='blup (inferred trees)', color=cmap(1), linestyle='--')
plt.plot(times, rmse_mean_inf, label='mle (inferred trees)', color=cmap(1))
plt.plot(times, rmse_mean_true, label='mle (true trees)', color=cmap(0))

plt.xlabel('generations ago', fontsize=16)
plt.ylabel('root mean squared error', fontsize=16)

plt.tick_params(axis='both',labelsize=12)

plt.legend(fontsize=12)
plt.savefig(fname)
plt.tight_layout()

plt.show()
```
