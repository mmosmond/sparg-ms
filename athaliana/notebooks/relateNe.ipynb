```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename='../data/cvi_africa_and1001.EVA_2020-10-20.vcf.gz_dip2Norm.vcf.b.gz.vcf_atcgOnly.vcf.gz_chr4.vcf_keep-samples.vcf_biallelic.vcf_imputed.vcf.gz.hap_polarized_relate_extracted.coal'
time = np.loadtxt(filename, skiprows=1, max_rows=1)
coal = np.loadtxt(filename, skiprows=2)
```

```python
plt.step(time, 0.5/coal[2:])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('generations ago')
plt.ylabel('effective population size $N_e$')
plt.show()
```
