```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import matplotlib.cm as cm
import pandas as pd
```

functions

```python
def map_locations(sample_locations, mle_locations=None, times=None,
                  lon_0='auto', lat_0='auto', height=500,
                  sample_alpha=0.1, mean_sample_location=False, sample_colors=None,
                  colorby='time', color='auto', node_colors=None, scale='log', 
                  lines=True, alpha=0.75, initial_marker='o', initial_marker_size=50, marker='o', 
                  CIs=False, Vs=None,
                  legend=False, legend_labels=None, 
                  title=None, title_pos=[0.5, 0.165],
                  colorbar=True, ticks=None, ticklabels=None,
                  figsize=(10,12), fname=None, dpi=180, return_image=False, show=True):
    """
    function to plot locations on maps
    """

    # center plot
    if lon_0 == 'auto':
        lon_0 = np.mean(sample_locations[:,1])
    if lat_0 == 'auto':
        lat_0 = np.mean(sample_locations[:,0])
    
    # set up figure
    fig = plt.figure(figsize=figsize)
    ax1 = fig.add_subplot(111, adjustable='box', aspect=1) #square
    bm1 = Basemap(projection='nsper', #projection
                  lon_0=lon_0, lat_0=lat_0, #center point
                  satellite_height=height*1000., #height in meters
                  resolution='l') #low?
    bm1.drawmapboundary(fill_color='lightsteelblue', zorder=0) #fill the globe with a blue color 
    bm1.fillcontinents(color='lightgray',lake_color='lightsteelblue', zorder=1) #continents gray, lakes blue
    bm1.drawcountries(color='gray', zorder=2) #country boundaries to help orient
    bm1.drawstates(color='gray', zorder=2) #state boundaries to help orient in NA
    cmap = plt.get_cmap('viridis')

    # plot locations of samples
#     ulocs, indices = np.unique(locs, axis=0, return_index = True) #unique locations only
    ulocs = sample_locations # all locations
    if sample_colors is None:
        sample_colors = ['k' for _ in ulocs]
    xi, yi = bm1(ulocs[:,1], ulocs[:,0]) #on the projection
    ax1.scatter(xi, yi, alpha=sample_alpha, color=sample_colors, zorder=3)
    # plot mean location of samples (eg to see if ancestor different from that)
    if mean_sample_location:
        xi, yi = bm1(np.mean(ulocs[:,1]), np.mean(ulocs[:,0])) #on the projection
        ax1.scatter(xi, yi, alpha=1, color='red', zorder=4)
    
    # plot predicted locations
    if mle_locations is not None:
        for l,locus in enumerate(mle_locations):
            for i,node in enumerate(locus):
                xs, ys = bm1(node[:,1], node[:,0]) #long and lat through time
                ntimes = len(node) #number of time points
                if colorby == 'time':
                    if color == 'auto':
                        if scale == 'log':
                            colors = [cmap(np.log(time)/np.log(times[-1])) for time in times]
                        else:
                            colors = [cmap(time/times[-1]) for time in times] #color for times
                    else:
                        colors = [color for _ in range(ntimes)]
                if colorby == 'node':
                    colors = [node_colors[i] for _ in range(ntimes)]
                if legend and l == 0:
                    ax1.scatter(xs[0], ys[0], color=colors[0], alpha=alpha, marker=initial_marker, s=initial_marker_size, zorder=4, label=legend_labels[i]) #plot first time point 
                else:
                    ax1.scatter(xs[0], ys[0], color=colors[0], alpha=alpha, marker=initial_marker, s=initial_marker_size, zorder=4)
                for j in range(1,ntimes):
                    if lines:
                        ax1.plot(xs[j-1:j+1], ys[j-1:j+1], color=colors[j], alpha=alpha, zorder=4+j) #plot lines connecting times
                    ax1.scatter(xs[j], ys[j], color=colors[j], alpha=alpha, marker=marker, zorder=4+j) #plot markers at time points
                    if CIs: #confidence intervals
                        width, height, theta = plot_cov_ellipse(Vs[l,i,j], [xs[j],ys[j]], nstd=1.96, fill=True, color='k', ax=ax1)
                        poly = bm1.ellipse(node[j,1], node[j,0], width, height, 100, edgecolor='k', alpha=0.5, fill=False)
                    
    # plot title
    if title is not None:
        plt.figtext(title_pos[0], title_pos[1], title, fontsize=14, ha='center')

    # color bar
    if colorbar:
        cbar = fig.colorbar(cm.ScalarMappable(norm=None, cmap=cmap), ax=ax1, orientation='horizontal', shrink=0.75, pad=0.025)
        cbar.set_ticks(ticks)
        cbar.set_ticklabels(ticklabels)
        cbar.set_label('years before present', fontsize=14)
        cbar.ax.tick_params(labelsize=14)

    if legend:
        plt.legend()

    plt.tight_layout()

    if fname is not None:
        plt.savefig(fname, dpi=dpi, bbox_inches = 'tight', pad_inches = 0)

    if return_image:
        # return the plot as an image array for making gifs
        fig.canvas.draw() # draw the canvas, cache the renderer
        image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        image  = image.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        return image    
        
    if show:
        plt.show()
    else:
        plt.close()
```

```python
def filter_samples(locations, group, clusters):

    if group == 'Eurasia':
        not_na = [i for i,j in enumerate(locations) if j[1]>-30] #not in NA (longitude greater than -30), and not missing (since the inequality is false when nan)
    elif group == 'Eurasia-no-Japan':
        not_na = [i for i,j in enumerate(locations) if j[1]>-30 and j[1]<120] #not NA, Japan, or nan
    else:
        print('error: group does not exist!')
        return

    # try to make clusters from pairwise differences  
    #remove = []
    #for cluster in clusters:
    #  remove.append(cluster[1:]) #remove all but first sample in each cluster (a bit of a rough approach)
    #remove = [i for j in remove for i in j] #flatten the list

    # just remove all samples
    remove = np.unique(clusters) #remove all samples that are in a nearly identical pair

    sample_ix = range(len(locations)+1) #index for all samples
    sample_ix_nonidentical = [i for i in sample_ix if i not in remove] #list of nonidentical sample idices
    keep = [i for i in sample_ix_nonidentical if i in not_na] #list of samples to keep

    return keep
```

```python
def load_data(node,twoNe,u,numiter,threshold,tCutoff,dCutoff,style,tsplits,group,misplaced=False):

    mle_locations = []
    for chromosome in chromosomes:

        fname = datadir + filename + '_chr%d_2Ne%d_u%s_numiter%d_threshold%.1f_tCutoff%d_dCutoff%d_group%s_tsplits%s_node%d_%s.npy' %(chromosome,twoNe,u,numiter,threshold,tCutoff,dCutoff,group,tsplits,node,style)        
        
        try:
            mle_locations.append(np.load(fname))
        except:
            print('missing chromosome %d' %chromosome)
              
    return mle_locations
```

parameters

```python
datadir = '../data/'
chromosomes = range(1,6)
twoNe = 170000
u = 7e-09
numiter = 5
threshold = 0.5
nloci = 1000
dCutoff = 1000
tCutoff = 10000
# times = range(0,1001,100)
times = np.logspace(0,3,10)
loci = range(1000)
plotdir = '../plots/'
group='Eurasia-no-Japan'
tsplits = [10,100,1000]
style='ancestor-locations'
filename='1001G'
```

load locations of all samples

```python
locations = np.loadtxt('../data/1001G.locs')
clusters = np.load(datadir + '1001G_MAF0.05_dCutoff%d_clusters.npy' %dCutoff, allow_pickle=True) #clusters of nearly identical samples
keep = filter_samples(locations, group, clusters)
```

load mles

```python
NA_nodes = [i for i,j in enumerate(locations) if j[1]<-30]

mean_mles = []
for node in NA_nodes:
    mle_locations = np.array(load_data(node,twoNe,u,numiter,threshold,tCutoff,dCutoff,style,tsplits,group))
    mean_mles.append(np.mean(mle_locations, axis=(0,1,3)))
mean_mles = np.array(mean_mles)
```

plot

```python
for i in [len(times)-1]: #go all the way to the final time
    fname = (plotdir + filename + '_NA-locations_tsplits%s.png' %(tsplits)).replace(' ','')
#     fname=None
    map_locations(sample_locations=locations[keep], mle_locations=np.array([mean_mles[:,0:i+1]]), times=times,
                  fname=fname, 
                  mean_sample_location=True,
                  initial_marker='x', marker='x',
                  ticks=[i/3 for i in range(4)], ticklabels=[1,10,100,1000], scale='log') 
```
